# DISTRIBUIÇÃO DE FREQUÊNCIAS

# 0.0 Importações

```r
library(dplyr)
library(sampling)
library(TeachingSampling)
```

```r
enem2019_tratado <- read.csv('enem2019_tratado.csv', sep=',')
```

```r
head(enem2019_tratado)
```

```r
enem_sorocaba <- enem2019_tratado %>% filter(NO_MUNICIPIO_RESIDENCIA=='Sorocaba')
```

# 1.0 Tabela de frequências absolutas

```r
freq_abs <- table(enem_sorocaba$NU_IDADE)
freq_abs_df <- data.frame(freq_abs)
freq_abs_df[order(freq_abs_df$Freq, decreasing = TRUE), ]
```

# 2.0 Frequência relativa

```r
freq_rel <- prop.table(freq_abs)
freq_rel_df <- data.frame(freq_rel)
freq_rel_df[order(freq_rel_df$Freq, decreasing = TRUE),]
```

# 3.0 Porcentagem da frequência relativa

```r
p_freq_rel <- 100 * prop.table(freq_rel)
p_freq_rel
```

# 4.0 Criar uma linha com o total

```r
freq_abs <- c(freq_abs, sum(freq_abs))
names(freq_abs)[54] <- "Total"
freq_abs
```

# 5.0 Juntando a frequência relativa e a frequência percentual

```r
freq_rel <- c(freq_rel,sum(freq_rel))
```

```r
p_freq_rel <- c(p_freq_rel, sum(p_freq_rel))
```

```r
# tabela final com todos os valores

tabela_final <- cbind(freq_abs,
                     freq_rel = round(freq_rel, digits=5),
                     p_freq_rel = round(p_freq_rel, digits=2))
```

```r
tabela_final
```

# 6.0 Construindo classes de frequências

```r
intervalo_classes <- seq(10,75,5)
intervalo_classes
```

```r
tabela_classes <- table(cut(enem_sorocaba$NU_IDADE, breaks=intervalo_classes, right=FALSE))
```

```r
tabela_classes <- data.frame(tabela_classes)
tabela_classes
```
